```python
import csv
import pandas as pd
import urllib
import os.path
```

```python
lore_url = 'https://vectorization.computer/AJAX/get_src.php'
out_path = '../kernels_lore/orig/'
```

```python
x = pd.read_csv('lore2.csv')
x.rename({'function': 'func'}, axis='columns', inplace=True)
```

```python
rows = x.to_dict(orient='records')
```

```python
for row in rows:
    fname = out_path + 'lore_' + row['id'] + '_' + str(row['line']) + '.c'
    
    if os.path.isfile(fname):
        continue
    else:
        print(row['id'])
    
    data = urllib.parse.urlencode(row)

    res = urllib.request.urlopen(lore_url + '?' + data)
    code = res.read()
    
    if len(code) > 1:
        with open(fname, 'w') as fout:
            fout.write(code.decode('utf-8'))
    else:
        print('empty: ', lore_url + '?' + data)
```
